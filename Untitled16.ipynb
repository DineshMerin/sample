```python
# Creating a scatter plot with detailed axis labels
import matplotlib.pyplot as plt
import numpy as np
# Generating synthetic data for the example
np.random.seed(0)
x = np.random.normal(50, 10, 100)  # Example bivariate data: x (Advertising Spend)
y = 0.5 * x + np.random.normal(0, 5, 100)  # Example bivariate data: y (Sales), with some noise

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', edgecolor='black')
plt.title('Figure 1.2: Example of a Scatter Plot for Bivariate Data')
plt.xlabel('Advertising Spend (₹)')
plt.ylabel('Sales (₹)')
plt.grid(True)

# Adding a caption below the plot
#plt.figtext(0.5, -0.1, "Figure 1.2: The scatter plot shows the relationship between Advertising Spend and Sales, illustrating how changes in advertising expenditure impact sales figures.", wrap=True, horizontalalignment='center', fontsize=12)

plt.show()
```
